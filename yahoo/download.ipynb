```python
import warnings
import yfinance as yf
warnings.simplefilter(action='ignore', category=FutureWarning)

# Get the data for the stock AAPL
df0 = yf.download(['AAPL', 'AMD', 'AMZN', 'BABA', 'BIDU', 'GOOG', 'IBM', 'JD', 'META', 'MSFT', 'NVDA', 'ORCL', 'PDD', 'UBER'], '2014-01-01')

# Import the plotting library
import matplotlib.pyplot as plt
%matplotlib inline

# Plot the close price of the AAPL
df0['Adj Close'].plot()
plt.show()
```

```python
df1 = df0['Adj Close']
print(df1.head())
print(df1.tail())
print(df1.shape)
print(df1.columns)
df1.plot()
plt.show()
```

```python
import pandas as pd
df_price = df1.loc[(df1.index >= '2023-01-01') & (df1.index < '2024-01-01')]
print(df_price.head())
print(df_price.tail())
print(df_price.shape)
print(df_price.columns)
df_price.to_csv('../data/adj_close_2023.csv')
df_price.plot()
plt.show()
```

```python
df_price = pd.read_csv('../data/adj_close_2023.csv', index_col='Date')
print(df_price.head())
df_ret = df_price.pct_change()*252
print(df_ret.head())
print(df_ret.tail())
df_ret.plot()
plt.show()
```

```python
plt.plot(df_ret['PDD'], label='PDD')
plt.plot(df_ret['META'], label='META')
plt.legend()
plt.show()
```

```python
df_price_diff = df_price.diff()
print(df_price_diff.head())
print(df_price_diff.tail())
df_price_diff.plot()
plt.show()
```
